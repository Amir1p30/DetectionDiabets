```python
#import all the library
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import  train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import  LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df
```

```python
#for preprocessing we need an array
arr_df=np.array(df)
```

```python
#division the input ans output 
X=arr_df[:,:8]
Y=arr_df[:,-1]
```

```python
#instead of 0 put the mean for precision 
Si=SimpleImputer(missing_values=0,strategy='mean')
X[:,1:5]=Si.fit_transform(X[:,1:5])
```

```python
#increase the precision  
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
#devision train and test for algorithms
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
for Mashin in (LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier(),KNeighborsClassifier(),GaussianNB()):
    Mashin.fit(x_train,y_train)
    result=Mashin.score(x_test,y_test)
    print(Mashin,result)
```
